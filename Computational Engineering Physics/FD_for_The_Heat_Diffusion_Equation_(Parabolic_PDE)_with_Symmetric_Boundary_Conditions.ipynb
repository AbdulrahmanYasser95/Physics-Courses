```python
import numpy as np
```

```python
class HeatDiffusionEquationSolution:
  def mirrorGrid(self, grid):
    print(f'Half grid: {grid}')
    if len(grid)%2:
      print('Odd grid')
      startIndex = -len(grid)
    else:
      print('Even grid')
      startIndex = -len(grid)+1
    print(f'Start index: {startIndex}')
    latterHalf = grid[startIndex::-1]
    print(f'Latter half: {latterHalf}')
    mirrorredGrid = np.concatenate(grid, latterHalf)
    return mirrorredGrid

  def __init__(self, numberOfPositionSteps, r):
    '''Initialise'''
    self.N = numberOfPositionSteps
    Dx = 1/self.N
    print(f'Position difference = {Dx}')
    Dt = r*Dx**2
    print(f'Time difference = {Dt}')

    initialPositionTemperatureGridSize = int(self.N/2+2)
    print(
        'Initial position-temperature grid size = '
        + f'{initialPositionTemperatureGridSize}'
    )
    self.currentPositionTemperatureGrid = np.full(
        initialPositionTemperatureGridSize,
        100
    )
    self.currentPositionTemperatureGrid[0] = 50

    completeCurrentPositionTemperatureGrid = self.mirrorGrid(
        currentPositionTemperatureGrid
    )
    completeCurrentPositionTemperatureGrid
    # print(
    #     'Initial position-temperature grid: '
    #     + f'{self.mirrorGrid(self.currentPositionTemperatureGrid)}'
    # )

  def getNextPositionTemperatureGrid(self):
    nextPositionTemperatureGrid = np.empty(self.N+1)
    for i in range(1, len(self.currentPositionTemperatureGrid)-1):
      print(i)

solution = HeatDiffusionEquationSolution(10, 0.5)
```
