```python
import numpy as np
```

```python
class HeatDiffusionEquationSolution:
  def __init__(
      self,
      leftSourceTemperature,
      rightSourceTemperature,
      initialTemperature,
      length,
      numberOfPositionSteps,
      r=0.5
  ):

    if r > 0.5:
      print('Warning: This solver uses the Euler method')
      print('Warning: r>0.5 will give unstable solutions')

    self.numberOfPositionSteps = numberOfPositionSteps
    self.r = r

    self.Dx = length/self.numberOfPositionSteps
    print(f'Position difference = {self.Dx}')
    self.Dt = self.r*self.Dx**2
    print(f'Time difference = {self.Dt}')
    self.t = 0

    self.rightSourceTemperature = float(rightSourceTemperature)
    self.leftSourceTemperature = float(leftSourceTemperature)

    self.currentPositionTemperatureGrid = np.full(
        numberOfPositionSteps+1,
        float(initialTemperature)
    )
    self.currentPositionTemperatureGrid[0] = float((
        leftSourceTemperature + initialTemperature
    )/2)
    self.currentPositionTemperatureGrid[-1] = float((
        rightSourceTemperature + initialTemperature
    )/2)
    print(
        'Initial position-temperature grid: '
        + f'{self.currentPositionTemperatureGrid}'
    )

    self.history = {0: self.currentPositionTemperatureGrid}

    print('Initialisation complete')
    print('------------------------------------------------------------------')

  def calculateNextPositionTemperatureGrid(self):
    nextPositionTemperatureGrid = np.copy(self.currentPositionTemperatureGrid)

    nextPositionTemperatureGrid[0] = self.leftSourceTemperature


    for i in range(1, self.numberOfPositionSteps):
        nextPositionTemperatureGrid[i] = float(
            self.r*(
              self.currentPositionTemperatureGrid[i+1] +\
              self.currentPositionTemperatureGrid[i-1]
            )
            + (1-2*self.r) * self.currentPositionTemperatureGrid[i]
      )

    nextPositionTemperatureGrid[-1] = self.rightSourceTemperature
    return nextPositionTemperatureGrid

  def advance(self, increments = 1):
    for i in range(increments):
      self.currentPositionTemperatureGrid = self.calculateNextPositionTemperatureGrid()
      self.history[self.Dt*i] = self.currentPositionTemperatureGrid

  def showHistory(self):
    # print(self.history)
    for t in self.history:
      print(f'{t}: {self.history[t]}')
```

# Cooling

## Δx = 0.2 & r= 0.25

```python
solution = HeatDiffusionEquationSolution(0, 0, 100, 1, 5, 0.25)
solution.advance(int(1/0.01))

solution.showHistory()
```

Φ(0.2,0.98) = 0.00368719

## Δx = 0.2 & r= 0.5

```python
solution = HeatDiffusionEquationSolution(0, 0, 100, 1, 5, 0.5)
solution.advance(int(1/0.02))

solution.showHistory()
```

Φ(0.2,0.98) = 0.00202227
