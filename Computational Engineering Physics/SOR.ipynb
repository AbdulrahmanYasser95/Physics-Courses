```python
import numpy as np
```

```python
def iterateWithSOR(initialGrid, unknownNodes, w, h, g, K):
  currentGrid = initialGrid.astype(float)
  for k in range(K):
    pastGrid = currentGrid
    # print(f'Past grid at iteration {k}:')
    # print(pastGrid)
    for unknownNode in unknownNodes:
      x, y = unknownNode
      # print(f'currentGrid[{x}][{y}] = {pastGrid[x][y]}'
      #   + f'+ ({w/4})*('
      #   + f'{pastGrid[x+1][y]} '
      #   + f'+ {currentGrid[x-1][y]} '
      #   + f'+ {currentGrid[x][y-1]} '
      #   + f'+ {pastGrid[x][y+1]} '
      #   + f'+ {-4*pastGrid[x][y]} '
      #   + f'-{h**2}*{g(x,y)}'
      #   + f') = {pastGrid[x][y] + (w/4)*(pastGrid[x+1][y]+ currentGrid[x-1][y]+ currentGrid[x][y-1]+ pastGrid[x][y+1] -4*pastGrid[x][y] -h**2*g(x,y))}'
      # )
      currentGrid[x][y] = pastGrid[x][y]\
        + (w/4)*(\
          pastGrid[x+1][y]\
          + currentGrid[x-1][y]\
          + currentGrid[x][y-1]\
          + pastGrid[x][y+1]\
          -4*pastGrid[x][y]\
          -g(x,y,h)\
        )
  print(f'Current grid at iteration {k}:')
  print(currentGrid)
```

##Analytical Method

```python
va1 = 0
va2 = 0
h = 1/4

# Maximum value without causing an overflow error (decreasing)
I = 20
for N in range(I):
  n = 2*N+1
  # print(n)
  va1 += ((4*np.sinh(h*np.pi*n)/(np.pi*n*np.sinh(np.pi*n))))*(
      30*np.sin(h*np.pi*n)
      # + 20*np.sin(h*np.pi*n)
      - 10*np.sin((1-h)*np.pi*n)
  )

for M in range(1,I+1):
  for L in range(1,I+1):
    # print(M, L)
    # A = (((-1)**M)/((M**2+L**2)*np.pi**2)) * 144/(M*L*np.pi)
    A = 36*((-1)**M)*(((-1)**L+1) - 2)/(M*L*np.pi)
    va2 += A*np.sin(np.pi*M*h)*np.sin(np.pi*L*h)

print(va1)
print(va2)
print(va1+va2)
```
