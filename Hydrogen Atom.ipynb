```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA

from ipywidgets import interact, widgets, interact_manual
from SchrodingerEquationSolver import SchrodingerEquationSolver
```

```python
# Can't extend the SE solver as it uses a cartesian laplacian
class HydrogenAtom(SchrodingerEquationSolver):
  """Extends the Schrodinger equation solver 
  to solve for the Hydrogen atom"""
  def __init__(self, m, Ngrid, xmin, xmax):
    super(HydrogenAtom, self).__init__(m, Ngrid, xmin, xmax)
    self.Vx = -1/self.xvec
```

```python
def SolveHydrogenAtom(m, Ngrid, xmin, xmax):
  """Solve the Schrodinger equation for the Hydrogen atom potential
  
  Args:
    m (float): Electron mass
    Ngrid (int): Number of grid points
    xmin (float): Minimum grid point
    xmax (float): Maximum grid point"""
  HA = HydrogenAtom(m, Ngrid, xmin, xmax)
  HA.solve()
```

```python
interact_manual(
  SolveHydrogenAtom, 
  m = widgets.FloatText(value = 1), 
  Ngrid = widgets.IntText(value = 1001), 
  xmin = widgets.FloatText(value = 1), 
  xmax = widgets.FloatText(value = 10)
)
```
