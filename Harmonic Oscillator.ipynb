```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg as LA

from ipywidgets import interact, widgets, interact_manual
from SchrodingerEquationSolver import SchrodingerEquationSolver
```

```python
class QuantumHarmonicOscillator(SchrodingerEquationSolver):
  """Extends the Schrodinger equation solver 
  to solve for a simple harmonic oscillator
  
  Args:
    omega (float): The simple harmonic oscillator's natural frequency"""
  def __init__(self, omega, m, Ngrid, xmin, xmax):
    super(QuantumHarmonicOscillator, self).__init__(m, Ngrid, xmin, xmax)
    self.omega = omega
    self.Vx = self.m * self.omega**2 / 2 * self.xvec**2
```

```python
def SolveQuantumHarmonicOscillator(omega, m, Ngrid, xmin, xmax):
  """Solve the Schrodinger equation for a simple harmonic oscillator potential
  
  Args:
    omega (float): The simple harmonic oscillator's natural frequency
    m (float): Particle mass
    Ngrid (int): Number of grid points
    xmin (float): Minimum grid point
    xmax (float): Maximum grid point"""
  QHO = QuantumHarmonicOscillator(omega, m, Ngrid, xmin, xmax)
  QHO.solve()
```

```python
interact_manual(
  SolveQuantumHarmonicOscillator, 
  omega = widgets.FloatText(value = 1), 
  m = widgets.FloatText(value = 1), 
  Ngrid = widgets.IntText(value = 1001),
  xmin = widgets.FloatText(value = -10), 
  xmax = widgets.FloatText(value = 10)
)
```
