```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Δt = .01
def get_N_timelines(λ):
    all_N = []
    for N0 in 10**np.arange(1,6):
        N = [N0]
        while True:
            number_decays = (np.random.random(size=(N[-1],)) < λ*Δt).sum()
            N.append(N[-1]-number_decays)
            if N[-1] == 0: break
        all_N.append(np.array(N))
    return all_N

def plot_N_timelines(all_N, λ):
    for i in range(5):
        t=np.arange(0,len(all_N[-1])*Δt)
        N0 = 10**(i+1)
        plt.semilogy(np.arange(0,len(all_N[i]))*Δt, all_N[i], label=f"$N_0=10^{i+1}$", color=f"C{i}")
        plt.semilogy(N0 * np.exp(-t*λ), label=f"$N_0=10^{i+1}$", alpha=.3)
    plt.ylim(.7,130000)
    plt.xlabel("Time t [1s]")
    plt.ylabel("N(t)")
    plt.title("Decay of N for different starting values")
    plt.legend()
```

```python
all_N = get_N_timelines(0.03)
plot_N_timelines(all_N, 0.03)
```

```python
all_N = get_N_timelines(0.3)
plot_N_timelines(all_N, 0.3)
```

```python
all_N = get_N_timelines(0.9)
plot_N_timelines(all_N, 0.9)
```
