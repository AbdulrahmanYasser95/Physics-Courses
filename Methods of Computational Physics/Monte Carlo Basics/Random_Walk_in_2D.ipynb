```python
import random
from math import sqrt
import matplotlib.pyplot as plt
import numpy as np
```

# a

```python
def take_random_steps(N):
  x = [0]
  y = [0]

  i = 0
  while i < N:

    deltaXPrime = random.random()*2-1
    # deltaXPrime

    deltaYPrime = random.random()*2-1
    # deltaYPrime

    L = sqrt(deltaXPrime**2 + deltaYPrime**2)
    # L

    if L > 0:
      deltaX = deltaXPrime/L
      # print('deltaX:', deltaX)
      deltaY = deltaYPrime/L
      # print('deltaY:', deltaY)

      x.append(x[i] + deltaX)
      y.append(y[i] + deltaY)

      i = i + 1

  # print(x)
  # print(y)

  return x, y
```

```python
fig = plt.figure(figsize=(7,7))
for i in range(3):
  x, y = take_random_steps(1000)
  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
```

# b

```python
def getRs(M, N):
  RSquared = []
  R = []
  for j in range(M):
    x, y = take_random_steps(N)
    RSquaredJ = x[N-1]**2 + y[N-1]**2
    RSquared.append(RSquaredJ)
    R.append(sqrt(RSquaredJ))

  averageRSquared = sum(RSquared)/M
  print('Actual Rrms: ', sqrt(averageRSquared))
  print('Theoretical Rrms: ', sqrt(N))

  return R, RSquared
```

```python
R, RSquared = getRs(100000, 10000)
```

```python
plt.hist(R)
plt.xlabel('R')
plt.ylabel('Density')
```
