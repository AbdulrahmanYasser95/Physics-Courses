```python
import matplotlib.pyplot as plt
import numpy as np
# plt.rcParams['text.usetex'] = False
```

```python
# Define the function
def polynomial_function(x, m, l):
    return m*x ** 2 + l*x ** 4

def polynomial_function_6(x, m, l):
    return m*x ** 2 + l*x ** 6

def polynomial_function_8(x, m, l):
    return m*x ** 2 + l*x ** 8

def polynomial_function_all(x, m, l):
    return m*x ** 2 + l*(x**4 + x**6 + x**8)

def polynomial_function_multiple_minima(x, m, l):
    return m*x ** 2 + (6.8*x**4 + -10*x**6 + +10*x**8 + -7.7*x**10 + 2.4*x**12)

def polynomial_function_odd(x, m, l):
  return m*x ** 2 + l*x**3 + x**4

def polynomial_function_odd_1m(x, m, l):
  return m*x ** 2 + 20*x**3 + 100**4

def eff_pot(x):
  return -x**2 +1.6*x**4*np.log(1.3*x**2) -2*x**4*np.log(0.8*x**2)

# Generate x values
x_values = np.linspace(-3, 3, 400)

m_p = 1
m_o = 0
m_n = -1
```

```python
l_o = 0

# Calculate y values
y_op = polynomial_function(x_values, m_p, l_o)
y_oo = polynomial_function(x_values, m_o, l_o)
y_on = polynomial_function(x_values, m_n, l_o)

# Plot the function
plt.plot(x_values, y_op, label=f'$μ^2$ > 0')
plt.plot(x_values, y_oo, label=f'$μ^2$ = 0')
plt.plot(x_values, y_on, label=f'$μ^2$ < 0')

# Add labels and title
plt.xlabel('ϕ')
plt.ylabel('V(ϕ)')
plt.title('V(ϕ) VS ϕ for λ = 0')
plt.xlim(-1,1)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
