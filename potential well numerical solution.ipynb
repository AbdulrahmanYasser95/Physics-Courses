```
timport numpy as np
import math

from math import sin
```

```
def fillHamiltonianMatrix(npw, kn, v0, a, b):
  H = np.zeros((npw,npw))
  for i in range(npw):
    for j in range(npw):
      if i==j:
        H[i][j] = kn[i]*kn[i] -v0*b / a
      else:
        deltaKn = kn[i] - kn[j]
        H[i][j] = -2* v0 * sin(deltaKn*b/2) / (a*deltaKn)
  return H
```

```
def fillKMatrix(npw, a):
  kn = np.zeros(npw)
  kn[0] = 0
  for i in range(1,npw,2):
    kn[i] = (i + 1) * math.pi / a
    kn[i+1] = -kn[i]
  return kn
```

```
n = 100
npw = 2*n + 1
# v0 = 0.0771605504587156
v0=1
# b = 2 * 85.06616257088847
b=2
# The value of a below gives the eigenvalue in the notes
a = 4.8*math.pi
# This value of a gives, approximately, the first exact energy with n = 100
# a= 100000*math.pi
# a = 2*math.pi
```

```
kn = fillKMatrix(npw, a)
H = fillHamiltonianMatrix(npw, kn, v0, a, b)
```

```
eigenvalues = np.linalg.eigvals(H)
eigenvalues.sort()
```

```
# The first 3 eigenvalues should be 0.008284444757973622, 0.01676595311625734 & 0.02486443843677008
print('Lowest eigenvalues:')
print(eigenvalues[0])
print(eigenvalues[1])
print(eigenvalues[2])

print('Highest eigenvalue:')
print(eigenvalues[npw-1])
```
